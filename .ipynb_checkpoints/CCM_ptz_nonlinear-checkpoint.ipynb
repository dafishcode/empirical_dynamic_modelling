```python
length = 800
max_list = list(range(len(sub_p20)))
for o in range(len(sub_p20)):
    mean = np.load(sub_p20[o], allow_pickle=True).item()['mean trace']

    #Apply sliding window over mean trace
    window = adfn.window(30, mean.shape[0])[0]
    diffs = list(range(mean.shape[0] - window))
    for t in range(mean.shape[0]):
        if t == mean.shape[0] - window:
            break
        diff_vec = np.zeros(window)
        for e in range(window):
            diff_vec[e] = mean[t+e] - mean[t] 
        diffs[t] = diff_vec
        
    sums, maxs = list(range(len(diffs))), list(range(len(diffs)))
    for i in range(len(diffs)):
        maxs[i] = np.max(diffs[i])
        
    if o == 0:
        max_list[o] = np.where(maxs == np.max(maxs[3000:]))[0]
        
    else:
        max_list[o] = np.where(maxs == np.max(maxs))[0]
    
    fig, ax = plt.subplots(figsize= (20,2))  
    plt.plot(mean, linewidth = 2, c= 'k')
    for x in np.arange(max_list[o], max_list[o]+length): plt.axvline(x=x, c='red',alpha = 0.05)
    plt.show()
```

```python
def name_zer(k):
    if len(str(k)) == 2:
        str_k = '00' + str(k)
        return(str_k)
        
    elif len(str(k)) == 3:
        str_k = '0' + str(k)

        return(str_k)
    return(str(k))

def remove(data):
    #Remove data with nothing in it
    store = []
    for i in range(len(data)):
        if sum(data[i]) == 0:
            store = np.append(store, i)

    new_data = np.zeros((data.shape[0] - len(store), data.shape[1]))
    
    
    if len(store) > 0:
        count= 0
        for i in range(len(data)):
            if sum(i == store) == 0:
                new_data[count] = data[i]
                count+=1
        return(new_data)
    
    else:
        return(data )
```

```python
#Cluster the neurons 

import h5py

for name in dic_l:
    data = np.load(name, allow_pickle=True).item()
    coord = data['coord']
    coord = coord[:,:3].astype(float)
    trace = data['trace']
    
    for k in n_clust:
        k_coord, k_trace = cfn.kspace_meantrace(coord, trace, k)
        mean_trace = np.apply_along_axis(np.mean, 0, k_trace)
        f_dict = { 'k coord': k_coord, 'k trace' : k_trace, 'mean k trace': mean_trace}
        
        strk = name_zer(k)
        
        save_dict = adfn.par_save_name(name, 'kmeans2-' + strk + '_')  + '_pre-CCM.npy'
        np.save(Fdata + save_dict, f_dict)

        save_h5 = cfn.par_save_name(name, 'kmeans2-' + strk + '_')  + '_trace_pre-CCM.h5'

        array = np.vstack((mean_trace, k_trace)).T
        f = h5py.File(save_h5, 'w')
        f.create_dataset("data", data = array)
        f.close()
    print('Done ' + name)
```

```python
num_l = '03', '06', '07', '10', '11', '13', '14', '15'
cond_l = 'BLN', 'PTZ05', 'PTZ20'
#n_clust = (np.geomspace(30,5000, 10)).astype(int)
datatype_l = 'mean rhodiff', 'mean e', 'mean correlation' ,'mean ccm', 'mean ccm neurons to seizure', 'mean ccm seizure to neurons'

#create dictionary to store it all
dic = {} 
for cond in cond_l: 
    dic.update({cond: {}})
    for typ in datatype_l: dic[cond].update({typ:np.zeros((len(num_l), len(n_clust)))})
        
num_l, cond_l, n_clust, datatype_l
```

```python
for x, num in enumerate(num_l):

    for y,cond in enumerate(cond_l):
        for z,n in enumerate(n_clust):
            data = adfn.return_files(Fdata, 'PTZ-WILDTYPE-CCM', '*' + num + '*' + cond + '*kmeans2*' + name_zer(n) + '_*xmap*')
            
            if len(data) == 0:
                continue
            
            elif len(data) > 1:
                print('More than one file found in search')
                break
            
            d = adfn.h5_2dict(data[0])

            #Extract
            data = mean_rd = np.mean(d['rhodiff'][1:, 1:]) #nonlinearity for each neuron vs every other neuron
            mean_e = np.mean(d['e'][1:]) #embedding dimension of each neuron
            mean_c = np.mean(d['rho'][1:, 1:]) #correlation of each neuron against all other neurons 
            mean_ccm = np.mean(d['ccm'][1:, 1:]) #ccm prediction of one neuron against all other neurons
            mean_c_to_sz = np.mean(d['ccm'][1:,0])#mean causation of neurons to seizure
            mean_sz_to_c = np.mean(d['ccm'][0,1:])#mean causation of seizure to neurons
            ext_l = mean_rd, mean_e, mean_c, mean_ccm, mean_c_to_sz, mean_sz_to_c

            #Store in dic
            for f,typ in enumerate(datatype_l):  
                dic[cond][typ][x][z] = ext_l[f]
                
    print(num)
np.save(Fdata + 'kmeans2_xmap.npy' , dic)
```
